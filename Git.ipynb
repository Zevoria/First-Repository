```python
<body>
    <h1>Lemon Spaghetti</h1>
    <p>Ingredients:</p>
    <ul>
        <li> 1 box spaghetti</li>
        <li> 2 lemons</li>
        <li> 1 cup parmesan cheese</li>
        <li> 1/2 cup olive oil</li>
        <li> fresh basil leaves</li>
    </ul>
    <p>Instructions:</p>
    <ol>
        <li>Cook spaghetti according to package directions.</li>
        <li> Zest and juice the lemons.</li>
        <li> In a large skillet, heat the olive oil over medium heat.</li>
        <li> Add the lemon juice and zest, and cook for 2 minutes.</li>
        <li> Stir in the parmesan cheese until melted.</li>
        <li> Drain the spaghetti and toss with the lemon sauce.</li>
        <li> Garnish with fresh basil leaves before serving.</li>
    </ol>
</body>
```
